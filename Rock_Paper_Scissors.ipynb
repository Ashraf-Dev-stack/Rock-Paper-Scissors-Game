```python
p1_score=0
p2_score=0
round_number=1
while round_number !=:
  print (f"\n--- Round {round_number} --- ")
  battle_code=input("Enter 2-Digit Battle Code (P1 P2):")

  #Input Validation
  if len(battle_code)!=2:
    print("Invalid input! You must exactly two keys.")
    continue

  #Speed Trap: Handling "10 vs "01"
  if  battle_code[0] in "123" and battle_code[1] in "890":
    p1_key=battle_code[0]
    p2_key=battle_code[1]
  elif battle_code[1] in '123' and battle_code[0] in '890':
    p1_key=battle_code[1]
    p2_key=battle_code[0]
  else:
    print("Invalid keys! P1 uses 1,2,3 and P2 uses 8,9,0.")
    continue

  #Translation Engine: Player 1
  if  p1_key == "1":
    p1_choice = "rock"
  elif p1_key == "2":
    p1_choice = "paper"
  else:
    p1_choice = "scissors"

  # Translation Engine: Player 2
  if p2_key == "8":
    p2_choice = "rock"
  elif p2_key == "9":
    p2_choice = "paper"
  else:
    p2_choice = "scissors"
  print(f"Player 1 choose {p1_choice}")
  print(f"Player 2 choose {p2_choice}")

  # Combat English
  if p1_choice == p2_choice:
    print("It's a tie!")
  elif p1_choice == "rock" and p2_choice == "scissors":
    print("Player 1 wins!")
    p1_score +=1
  elif p1_choice == "scissors" and p2_choice == "paper":
    print("Player 1 wins!")
    p1_score +=1
  elif p1_choice == "paper" and p2_choice == "rock":
    print("Player 1 wins!")
    p1_score +=1
  else:
    print("Player 2 wins!")
    p2_score += 1

  # Advance the round!
  round_number += 1
```
